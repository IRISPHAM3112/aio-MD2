```python
#Chuyen anh mau thanh anh den trang theo cach Lightness
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
img = mpimg.imread('/content/tải xuống.jpeg')
# plt.imshow("/content/tải xuống.jpeg")
r,g,b = img[:,:,0], img[:,:,1], img[:,:,2]
value_max = np.maximum(np.maximum(r,g), b)
value_min = np.minimum(np.minimum(r,g), b)
lightness = (value_max + value_min)/2

img_gray1 = img.copy()
img_gray1[:,:,0] = lightness
img_gray1[:,:,1] = lightness
img_gray1[:,:,2] = lightness

#mpimg.imsave('/content/tải xuống.jpeg', img_gray )
plt.imshow(img_gray1)
plt.axis('off')

plt.show()
#Chuyen anh mau sang anh trang den theo cach Average
r_avg = (r+g+b)/3
img_gray2 = img.copy()
img_gray2[:,:,0] = r_avg
img_gray2[:,:,1] = r_avg
img_gray2[:,:,2] = r_avg
plt.imshow(img_gray2)
plt.axis('off')
#Chuyen anh mau sang anh trang den theo cach Luminosity
img_gray3 = img.copy()
luminosity = 0.21*r + 0.72*g + 0.07*b
img_gray3[:,:,0] = luminosity
img_gray3[:,:,1] = luminosity
img_gray3[:,:,2] = luminosity
plt.imshow(img_gray3)
plt.axis('off')
```
